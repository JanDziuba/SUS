Load data

```python
import pandas as pd

car_table = pd.read_csv("car_table", sep=" ")

car_table
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

car_table = car_table.apply(le.fit_transform)

car_table
```

```python
from sklearn.model_selection import train_test_split 


X_train, X_test, y_train, y_test = train_test_split(car_table.loc[:, car_table.columns != "mpg"],
      car_table["mpg"], test_size=0.3, random_state=49) 
  
X_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier (criterion='gini')
tree_clf.fit(X_train, y_train)

print("Test score: {:.2f}".format(tree_clf.score(X_test, y_test))) 
```

```python
from sklearn.metrics import confusion_matrix

y_pred = tree_clf.predict ( X_test )
confusion_matrix ( y_test , y_pred )
```
